```python
from nilearn.datasets import MNI152_FILE_PATH

print(f"Path to template: {MNI152_FILE_PATH!r}")
```

```python
from nilearn import plotting

plotting.plot_img(MNI152_FILE_PATH)
```

## Image maniplulation

```python
# Smoothing
from nilearn import image
smoothed_img = image.smooth_img(MNI152_FILE_PATH, fwhm=8)
plotting.plot_img(smoothed_img)

another_smoothed_img = image.smooth_img(smoothed_img, fwhm=4)
plotting.plot_img(another_smoothed_img)
```
